```python
webtoon = pd.read_csv('webtoons_data.csv')
```

```python
# Want to convert strings of the genre column into integers
def gen_to_int(word):
    genre = np.array(webtoon['Genre'].unique())
    for i in range(len(genre)):
        if word == genre[i]:
            return i+1
webtoon['Genre'] = webtoon['Genre'].map(gen_to_int)

# Want to convert strings of the status column into integers
def stat_to_int(word):
    status = np.array(["MONDAY","TUESDAY","WEDNESDAY","THURSDAY","FRIDAY","SATURDAY","SUNDAY","COMPLETED"])
    for i in range(len(status)):
        if word == status[i]:
            return i+1
webtoon['Status'] = webtoon['Status'].map(stat_to_int)
```

```python
wt_data = np.loadtxt('wt_model.csv', delimiter=',', skiprows = 1, usecols=(1,2,3,4,5))
```

```python
X = np.zeros([wt_data.shape[0],4])
X.shape
```

```python
X[:,:2] = wt_data[:,0:2]
X[:,2:] = wt_data[:,3:]
# Normalize every feature and target value by subtracting their means
X = X - np.mean(X,axis=0) 
y = wt_data[:,2]
y = y - np.mean(y)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

Linear Regression

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state = 0)
wine_reg = LinearRegression()
wine_reg.fit(X_train,y_train)
y_pred = wine_reg.predict(X_test)
y_pred_train = wine_reg.predict(X_train)
print("MSE is for the training data:", np.mean((y_pred_train - y_train)**2) )
print("MSE is for the testing data:", np.mean((y_pred - y_test)**2) )
print("R squared for training data is:", wine_reg.score(X_train,y_train))
print("R squared for testing data is:", wine_reg.score(X_test,y_test))
```

KMeans
